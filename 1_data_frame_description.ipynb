Create a subset of the data satifying the following conditions 
# 1. All the data for the year 2022 only.

```python
covid_df['date'] = pd.to_datetime(covid_df['date'])  #pd.to_datetime convert the date to datetime objects
covid_df[covid_df['date'].dt.year==2022]
```

2. Denmark, but only the data from 2020. 

```python

covid_df[(covid_df['location_key']=='DK') & (covid_df['date'].dt.year==2020)]
```

3. India, but also all the dates where the new cases reported were more than 1000. 

```python
covid_df[(covid_df['location_key']=='IN') & (covid_df['new_confirmed']>1000)]
```

4. Subset the data to only include the following columns: "date","location_key","new_confirmed", "new_deceased", "new_recovered", "new_tested"

```python
covid_df.iloc[:,:6]
```

6. Add a new column to the dataset: The column should be named as "year" and should only contain the year part of the date, and column should be named as "day" and should only contain the day of the week for each row.

```python
covid_df['year']=covid_df['date'].dt.year
covid_df['Day']=covid_df['date'].dt.day_of_week
covid_df
```

Find the number of recovered patients in each location in the year 2021
#groupby- This can be used to group large amounts of data and compute operations on these groups.

```python
recovered_2021=covid_df[(covid_df['date'].dt.year==2021) & (covid_df['new_recovered'])]
recovered_2021.groupby(["year","location_key"])['new_recovered'].sum()
```

Find mean day of the week

```python
mean_dayofweek=covid_df.groupby(["Day","location_key"])['new_tested'].mean()
mean_dayofweek
```

Find the mortality rate, MR=new deceased/new confirmed

```python
covid_df['mortality_rate']=covid_df['new_deceased']/covid_df['new_confirmed']
covid_df.sample(3)
```
